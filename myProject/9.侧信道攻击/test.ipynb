### collections.defaultdict

collections.defaultdict 在 Python 中是一个字典的子类，当访问不存在的键时，它会自动提供一个默认值。这消除了显式检查（例如 if key not in dict ）的需要，并防止了 KeyError 异常。

```python
from collections import defaultdict
d = defaultdict(list)  # 默认值为列表

d['fruits'].append('apple')
d['vegetables'].append('carrot')

print(d)
print(d['juices'])
```

### time.perf_counter_ns()

```python
from time import perf_counter_ns

# 输入n和m，n表示输入的次数，m表示要判断的数
n, m = map(int, input('请输入n和m，用空格隔开：').split())  #! 这里涉及map函数的用法

# 开始测量时间
start = perf_counter_ns()

for i in range(n):
    t = int(input('请输入一个数：'))  # 输入一个数
    if t % m == 0:
        print(t)

# 结束测量时间
stop = perf_counter_ns()

print("开始测量时间:", start, 'ns')
print("结束测量时间:", stop, 'ns')

print("Elapsed time during the whole program",
      stop-start, 'ns')

print('换算成秒为：', (stop-start)/1e9)
```

```python
from scipy import stats
import numpy as np

A = np.array([78,84,92,88,75,80,85,90,87,7978,84,92,88,75,80,85,90,87,79])
B = np.array([82,88,75,90,78,85,88,77,92,8082,88,75,90,78,85,88,77,92,80])

t_val, p_val = stats.ttest_ind(A, B)  # 独立样本T检验公式

alpha = 0.05  # P临界值
df = len(A)+len(B)-2  # 自由度

crit_t = stats.t.ppf(1 - alpha/2, df)  # 临界t值

print("T-value:", t_val)
print("P-Value:", p_val)
print("Critical t-value:", crit_t)

print('T-test Result:')
if np.abs(t_val) >crit_t:
    print('Significant difference found.')
else:
    print('No significant difference.')

print('P-test Result:')
if p_val >alpha:
    print('Fail to reject H0. No strong evidence of difference.')
else:
    print('Reject H0. Significant difference found.')
```

通过 threading.Lock()保证线程同步

创建锁：lock = threading.Lock()

锁定和释放：lock.acquire()和lock.release()

```python
import datetime
import os
import threading
import time


def log(msg):
    pid = os.getpid()
    tid = threading.current_thread().ident
    print(f"进程:[{pid}]线程:[{tid}]{msg}")


def add(lock, x, y):
    lock.acquire()  # 加锁，确保线程同步
    now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    log(f"现在的时间是:{now},开始加法运算")
    log(f"执行加法:{x} + {y} = {x + y}")
    time.sleep(3)
    now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    log(f"现在的时间是:{now},结束加法运算")
    lock.release()  # 解锁，允许其他线程继续执行
    return x + y


if __name__ == '__main__':
    thread_lock = threading.Lock()  # 创建锁，用于线程同步
    t1 = threading.Thread(target=add, args=(thread_lock, 1, 2))
    t2 = threading.Thread(target=add, args=(thread_lock, 3, 4))
    t1.start()
    t2.start()
```

这样就保证了线程同步。函数中的块代码是一起执行的。

```python
import datetime
import os
import threading
import time


def log(msg):
    pid = os.getpid()
    tid = threading.current_thread().ident
    print(f"进程:[{pid}]线程:[{tid}]{msg}")


def add(lock, x, y):
    lock.acquire()  # 获得锁，阻塞其他线程
    now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    log(f"现在的时间是:{now},开始加法运算")
    log(f"执行加法:{x} + {y} = {x + y}")
    time.sleep(3)
    now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    log(f"现在的时间是:{now},结束加法运算")
    lock.release()  # 释放锁，允许其他线程继续执行
    return x + y


if __name__ == '__main__':
    log(f'这是main线程的开始:{datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")}')
    thread_lock = threading.Lock()  # 创建一个锁对象
    t1 = threading.Thread(target=add, args=(thread_lock, 1, 2))
    t2 = threading.Thread(target=add, args=(thread_lock, 3, 4))
    t1.start()
    t2.start()
    t1.join()  # 等待t1执行结束再执行后面的语句
    t2.join()  # 等待t2执行结束再执行后面的语句
    log(f'这是main线程的结束:{datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")}')
```

通过daemon设置守护线程

当一个进程中的主线程和其他非守护线程都结束时，则守护线程也会随着他们的结束而结束，不再执行后续代码

换句话说，如果一个进程中还存在主线程或者还存在非守护线程，则守护线程自己没执行完自己时，就还会继续存在，继续执行自己的代码。

举几个例子说明：

（1）加法线程执行2秒，减法线程执行6秒，主线程执行4秒，都是非守护线程的时候，则都正常执行完毕后程序才会退出

```python
import datetime
import os
import threading
import time


def log(msg):
    pid = os.getpid()
    t = threading.current_thread()
    print(f"进程:[{pid}]线程:[{t.ident}-{t.name}]{msg}")


def add(lock, x, y):
    lock.acquire()
    now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    log(f"现在的时间是:{now},开始加法运算")
    log(f"执行加法:{x} + {y} = {x + y}")
    time.sleep(2)
    now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    log(f"现在的时间是:{now},结束加法运算")
    lock.release()
    return x + y


def sub(lock, x, y):
    lock.acquire()
    now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    log(f"现在的时间是:{now},开始减法运算")
    log(f"执行减法:{x} - {y} = {x - y}")
    time.sleep(6)
    now = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    log(f"现在的时间是:{now},结束减法运算")
    lock.release()
    return x - y


if __name__ == '__main__':
    log(f'这是main线程的开始:{datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")}')
    thread_lock = threading.Lock()
    t1 = threading.Thread(target=add, args=(thread_lock, 1, 2), name="", daemon=False)  # 不指定名称默认线程名称为Thread-[数字]
    t2 = threading.Thread(target=sub, args=(thread_lock, 8, 4), name='', daemon=False)  # 不指定名称默认线程名称为Thread-[数字]
    t1.start()
    t2.start()

    sleep = 4
    log(f"main在睡眠{sleep}秒")
    time.sleep(sleep)
    log(f'这是main线程的结束:{datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")}')
```

大家都按照自己的执行需求完成了执行，加法线程执行2秒，减法线程执行6秒，主线程执行4秒

（2）加法线程执行2秒，减法线程执行6秒，主线程执行4秒，设置减法线程为守护线程，则预期情况是加法线程完成执行2秒，主线程执行4秒，减法线程执行一些代码，然后后续丢失一些代码
