#### 1.public_sst_data.iloc[start_idx:end_idx]
    pandas.DataFrame.iloc[]
    iloc 是 pandas DataFrame 的整数位置索引方法
    它通过行号（从0开始的整数）来选取数据
    iloc[start:end] 选取从 start 行到 end-1 行的数据（左闭右开区间）

```python
import pandas as pd

mydict = [
    {'a': 1, 'b': 2, 'c': 3, 'd': 4},
    {'a': 100, 'b': 200, 'c': 300, 'd': 400},
    {'a': 1000, 'b': 2000, 'c': 3000, 'd': 4000}]

df = pd.DataFrame(mydict)  #  key 为列名, value 为行数据
df
```

```python
df.iloc[0]
```

```python
df.iloc[0]['a']  # 第0行的a列
```

```python
df.iloc[1]  # 第1行
```

```python
df.iloc[0:2]  # 第0行到第1行
```

#### 2.pandas.DataFrame.iterrows()
    iterrows() 是 DataFrame 的迭代器方法
    它按行遍历 DataFrame，每次返回一个包含行索引和行数据的元组
    语法：for index, row in dataframe.iterrows():

```python
mydict = [
    {'a': 1, 'b': 2, 'c': 3, 'd': 4},
    {'a': 100, 'b': 200, 'c': 300, 'd': 400},
    {'a': 1000, 'b': 2000, 'c': 3000, 'd': 4000}]

df = pd.DataFrame(mydict)  #  key 为列名, value 为行数据
df
```

```python
for index, row in df.iterrows():
    print(f"第 {index} 行")
    print(row)
    print("------------------")
```
