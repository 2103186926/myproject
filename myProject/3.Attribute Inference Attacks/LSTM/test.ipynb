### UTC（协调世界时）定义
UTC = Coordinated Universal Time（协调世界时）

    世界标准时间：全球统一的时间标准
    基准时区：位于英国伦敦的格林威治天文台（经度0°）
    无夏令时：UTC没有夏令时调整
    科学和技术标准：航空、天文、互联网等领域的通用时间

UTC与其他时间的关系:

    UTC = 格林威治标准时间（GMT）
    北京时间 = UTC + 8小时
    纽约时间 = UTC - 5小时（标准时间）或 -4小时（夏令时）
    伦敦时间 = UTC + 0小时（标准时间）或 +1小时（夏令时）

```python
### 为什么使用UTC？
# 问题：不同时区的服务器时间不一致
server_ny = "2024-01-15 10:30:00"  # 纽约时间
server_cn = "2024-01-15 23:30:00"  # 北京时间

# 解决方案：统一使用UTC
server_ny_utc = "2024-01-15 15:30:00+00:00"
server_cn_utc = "2024-01-15 15:30:00+00:00"
# 现在可以准确比较时间了！
```

#### 如何选择合适的时区？
    选择UTC的情况：
    服务器时间戳、数据库存储、API响应、日志文件、跨时区计算

    选择本地时区的情况：
    用户界面显示、报表生成、本地文件命名、面向最终用户的功能

#### 记住这个原则：
    "存储用UTC，显示用本地" - 这是处理时区问题的黄金法则！

### datetime.datetime 基本用法详解

#### 1.datetime.now(timezone.utc).isoformat()这行代码的解析
    datetime.now(timezone.utc)：获取当前的UTC时间。这里使用了timezone.utc作为时区参数，确保获得的时间是UTC时间，而不是本地时间。
    .isoformat()：将datetime对象转换为ISO 8601格式的字符串。这种格式的字符串通常类似于：'2023-10-05T12:34:56.789012+00:00'

```python
from datetime import datetime, timezone

# 获取当前日期和时间
now = datetime.now()
print("当前时间:", now)

# 获取当前UTC时间
current_utc = datetime.now(timezone.utc)
print("当前UTC时间:", current_utc)

# 转换为ISO 8601标准格式
iso_string = current_utc.isoformat()
print("ISO 8601格式:", iso_string)

# 等价于一行代码：
result = datetime.now(timezone.utc).isoformat()
print("等价一行代码结果:", result)
```

#### 获取当前日期和时间
我们可以使用 datetime 类的 now() 方法来获取当前的日期和时间。

```python
# 本地当前时间（无时区信息）
now_local = datetime.now()
print(now_local)  # 2024-01-15 10:30:45.123456

# UTC当前时间（带时区信息）
now_utc = datetime.now(timezone.utc)
print(now_utc)    # 2024-01-15 10:30:45.123456+00:00

# 只获取日期部分
today = datetime.today()
print(today)      # 2024-01-15 10:30:45.123456
```

#### 创建特定的日期和时间
我们可以通过 datetime 类的构造函数来创建特定的日期和时间。

```python
from datetime import datetime

# 创建特定的日期和时间
specific_time = datetime(2025, 5, 20, 13, 14, 30)
print("特定时间:", specific_time)
```

```python
# 方法1: 直接创建
dt1 = datetime(2024, 1, 15)           # 2024-01-15 00:00:00
dt2 = datetime(2024, 1, 15, 10, 30)   # 2024-01-15 10:30:00
dt3 = datetime(2024, 1, 15, 10, 30, 45)  # 2024-01-15 10:30:45
dt4 = datetime(2024, 1, 15, 10, 30, 45, 123456)  # 2024-01-15 10:30:45.123456

# 方法2: 从字符串解析
dt5 = datetime.strptime("2024-01-15", "%Y-%m-%d")
dt6 = datetime.strptime("15/01/2024 10:30", "%d/%m/%Y %H:%M")
```

```python
print(dt1)
print(dt2)
print(dt3)
print(dt4)
print(dt5)
print(dt6)
```

#### 格式化日期和时间
datetime 对象可以通过 strftime() 方法格式化为字符串。

```python
from datetime import datetime

# 获取当前时间
now = datetime.now()
print("当前时间:", now)

# 格式化输出
formatted_time = now.strftime("%Y-%m-%d %H:%M:%S")
print("格式化时间:", formatted_time)
```

```python
now = datetime.now()

# ISO格式（国际标准）
iso_format = now.isoformat()
print(iso_format)  # 2024-01-15T10:30:45.123456

# 自定义格式
custom_format = now.strftime("%Y-%m-%d %H:%M:%S")
print(custom_format)  # 2024-01-15 10:30:45

# 常用格式符号：
# %Y - 4位数年份 (2024)
# %m - 月份 (01-12)
# %d - 日期 (01-31)
# %H - 24小时制小时 (00-23)
# %M - 分钟 (00-59)
# %S - 秒 (00-59)
# %f - 微秒 (000000-999999)
```

#### 计算时间差
timedelta 类可以用于计算两个日期或时间之间的加减。

```python
from datetime import datetime, timedelta

# 获取当前时间
now = datetime.now()

# 计算 10 天后的时间
future_time = now + timedelta(days=10)
print("10 天后的时间:", future_time)
```

```python
from datetime import timedelta

now = datetime.now()

# 未来时间
future = now + timedelta(days=7)        # 7天后
future2 = now + timedelta(hours=3)      # 3小时后
future3 = now + timedelta(weeks=2, days=1, hours=5)  # 组合

# 过去时间
past = now - timedelta(days=1)          # 1天前

# 时间差计算
time_diff = future - now
print(time_diff.days)      # 天数差

time_diff = future2 - now
print(time_diff.seconds)   # 秒数差（不包括天数）
```

#### 常见应用场景
1.计算两个日期之间的天数

```python
from datetime import date

# 创建两个日期
date2 = date(2025, 11, 11)
date1 = date(2025, 9, 26)

# 计算天数差
delta = date2 - date1
print("两个日期之间的天数差:", delta.days)
```

#### 处理时区
datetime 模块本身不直接支持时区操作，但可以通过 pytz 库来处理时区。

```python
from datetime import datetime
import pytz

# 获取当前时间并设置时区
now = datetime.now(pytz.timezone('Asia/Shanghai'))
print("上海当前时间:", now)
```

```python
from datetime import timezone, timedelta

# UTC时间
utc_time = datetime.now(timezone.utc)

# 创建特定时区（如东八区）
beijing_tz = timezone(timedelta(hours=8))
beijing_time = datetime.now(beijing_tz)

# 时区转换
utc_time = datetime.now(timezone.utc)
local_time = utc_time.astimezone()  # 转换为本地时区

print(f"UTC: {utc_time}")
print(f"Beijing: {beijing_time}")
print(f"Local: {local_time}")
```

#### 提取时间组件

```python
now = datetime.now()

print(now.year)        # 年
print(now.month)       # 月
print(now.day)         # 日
print(now.hour)        # 时
print(now.minute)      # 分
print(now.second)      # 秒
print(now.microsecond) # 微秒
print(now.weekday())   # (0=周一, 6=周日)
print(now.isoweekday()) #(1=周一, 7=周日)
```
